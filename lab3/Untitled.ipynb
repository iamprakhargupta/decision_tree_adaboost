```python
c=500
# df=pd.DataFrame()
l=[]
i=0
browser.get('https://en.wikipedia.org/wiki/Main_Page')
while True:
    if i<=c:
        content_element = browser.find_element_by_id('n-randompage').find_element_by_css_selector('a')
        content_element.click()
        text=browser.find_element_by_id('mw-content-text').find_element_by_css_selector('p').text
        if len(text)>15:
            l.append(text)
            i+=1
    else:
        break
```

```python
import pickle
```

```python
file = open('english500.pkl', 'wb')
pickle.dump(l, file)
file.close()
```

```python
filehandler = open('english500.pkl', 'rb')
english = pickle.load(filehandler)
filehandler.close()
```

```python
for i in english:
    print(i)
```

```python
import re
l=[]
for i in english:
    s=re.sub(r'\(.*\)', '', i)
    s=re.sub(r'\[.\]', '', s)
    s=re.sub(r'\\n', '', s)
    s=re.sub(r'\\n', '', s)
    l.append(s)
    
```

```python
filehandler = open('dutch500.pkl', 'rb')
dutch = pickle.load(filehandler)
filehandler.close()
```

```python
d=[]
for i in dutch:
    s=re.sub(r'\(.*\)', '', i)
    s=re.sub(r'\[.\]', '', s)
    s=re.sub(r'\\n', '', s)
    d.append(s)
    
```

```python
english_sent=[]
for i in l:
    i=i.split(" ")
    x=len(i)
    g=x//15
    if g==0:
        continue
    else:
        for j in range(0,g):
            s=i[15*j:15*j+15]
            if len(s)!=15 or "" in s:
                continue
            english_sent.append(s)
```

```python
english_sent
en_sent=[]
for i in english_sent:
    en_sent.append("en|"+" ".join(i))
```
